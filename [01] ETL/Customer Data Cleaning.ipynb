```python
import pandas as pd
import numpy as np
```

```python
cust = pd.read_csv(r'C:\Users\Karen Fernandes\anaconda3\Files\Global Electronics Retailer\Customers.csv', encoding='latin1')
cust.head()
```

```python
cust.info()
```

```python
# converting the column names to lowercase 
cust.columns = cust.columns.str.lower()
```

```python
# convert the column names to snake case
cust.columns = cust.columns.str.lower().str.replace(' ', '_')
```

```python
# rename customer key column
cust.rename(columns={'customerkey': 'customer_key'}, inplace=True)
```

```python
# convert birthday to datetime
cust['birthday'] = pd.to_datetime(cust['birthday'])
```

```python
# checking for nulls
cust.isnull().sum()
```

```python
null_state_code_rows = cust[cust['state_code'].isnull()]
null_state_code_rows
```

```python
# state_code for Napoli is NA
# https://www.busuu.com/en/italian/abbreviations
cust['state_code'] = cust['state_code'].fillna('NA')
```

```python
# convert the 'city' column to proper case
cust['city'] = cust['city'].str.title()
```

```python
# checking for duplicates
duplicates = cust[cust.duplicated(subset='customer_key', keep=False)]
duplicates
```
